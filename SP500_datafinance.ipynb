```python
# install the following into your anaconda prompt >> pip install pandas matplotlib yfinance

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Set the start and end dates for the data
start_date = "2013-01-01"
end_date = "2023-01-01"

# Fetch S&P 500 historical data
sp500 = yf.download('^GSPC', start=start_date, end=end_date)

# Calculate daily returns
sp500['Daily_Return'] = sp500['Adj Close'].pct_change()

# Drop missing values from the first row
sp500.dropna(inplace=True)

# Plot the histogram of daily returns
plt.figure(figsize=(10, 6))
plt.hist(sp500['Daily_Return'], bins=100, alpha=0.7, color='b', edgecolor='k', label='S&P 500 Returns')
plt.xlabel('Daily Returns')
plt.ylabel('Frequency')
plt.title('S&P 500 Daily Returns Histogram (2013-2023)')
plt.legend()
plt.grid(True)
plt.show()

```
